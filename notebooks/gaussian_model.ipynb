# Example of GaussianModel

```python
import numpy as np
import matplotlib.pyplot as plt

from alea.examples import GaussianModel
```

```python
parameter_definition = {
    'mu': {
        'fit_guess': 0.,
        'fittable': True,
        'nominal_value': 0.,
        'parameter_interval_bounds': [
            -10,
            10,
        ],
    },
    'sigma': {
        'fittable': False,
        'nominal_value': 1.,
    },
}

model = GaussianModel(parameter_definition=parameter_definition)
```

```python
model.data = model.generate_data(mu=0, sigma=2)
best_fit, lf = model.fit(sigma=2)
```

```python
model.data[0]['hat_mu'] = 1.
mu = np.linspace(-3, 3, 101)
plt.plot(mu, model.ll(mu=mu))
plt.xlabel('mu')
plt.ylabel('likelihood')
plt.show()
```
